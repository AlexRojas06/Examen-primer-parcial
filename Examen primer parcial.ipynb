```python
def quitar_espacios(entrada_funcion):
    n=0
    bandera=False
    frase_aux=[]
    
    frase=list(entrada_funcion)
    if frase[-1]!='$':
        frase.append('$')  

    for i in frase:
        if i!=' ':
            frase_aux.append(i)
            bandera=True
        else:
            if bandera==True and frase[n+1]!='$' and frase[n+1]!=' ':
                frase_aux.append(i)
        n=n+1

    bandera=False
    
    for i in frase_aux:
        if bandera==False:
            frase_output=i
            bandera=True
        else:
            frase_output=frase_output+i
            
    return frase_output
    
```

```python
def separar_signos(entrada_funcion2):
    n=0
    bandera=False
    frase=list(entrada_funcion2)
    
    for i in frase:
        if i=='+' or i=='-':
            frase[n]=" "+i+" " 
        n=n+1    

    for i in frase:
        if bandera==False:
            frase_output=i
            bandera=True
        else:
            frase_output=frase_output+i
            
    return frase_output
```

```python
def separar_punto(entrada_funcion3):
    Numeros="1234567890"
    n=0
    bandera=False
    bandera1=False
    bandera2=False
    frase=list(entrada_funcion3)
    
    for i in frase:
        if i=='.':   
            for j in Numeros:                
                if j==frase[n-1]:
                    bandera1=True
                    
                if j==frase[n+1]:
                    bandera2=True
                                   
            if bandera1==True and bandera2==False:
                frase[n]=frase[n]+" "
            elif bandera1==False and bandera2==True:
                frase[n]=" "+frase[n]
            elif bandera1==False and bandera2==False:
                frase[n]=" "+frase[n]+" "   
            
            bandera1=False
            bandera2=False
        n=n+1 
        
    for i in frase:
        if bandera==False:
            frase_output=i
            bandera=True
        else:
            frase_output=frase_output+i
            
    return frase_output
```

```python
def separar_LetraNumero(entrada_funcion4):
    Numeros="1234567890"
    Abecedario="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    frase=list(entrada_funcion4)
    n=0
    bandera=False
    
    for i in frase:
        for j in Numeros:
            if j==i:
                for k in Abecedario:
                    if k==frase[n+1]:
                        frase[n]=frase[n]+" "
        n=n+1
    
    for i in frase:
        if bandera==False:
            frase_output=i
            bandera=True
        else:
            frase_output=frase_output+i
            
    return frase_output
```

```python
def separar_CaracterDesconocido(entrada_funcion5):
    Numeros="1234567890"
    Abecedario="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    frase=list(entrada_funcion5)
    n=0
    bandera=False
    bandera2=False
    
    for i in frase:
        for j in Numeros:
            if j==i:
                bandera=True
        for k in Abecedario:
            if k==i:
                bandera=True
        if i==' ' or i=='+' or i=='-' or i=='.' or i=='$':
            bandera=True
        
        if bandera==False:
            frase[n]=' '+frase[n]+' '
        
        bandera=False
        n=n+1
    
    for i in frase:
        if bandera2==False:
            frase_output=i
            bandera2=True
        else:
            frase_output=frase_output+i
            
    return frase_output
```

```python
def construir_lista(entrada_funcion6):
    frase=list(entrada_funcion6)
    n=0
    lista=[]
    bandera=False
    
    for i in frase:
        if i!=' ' and i!='$':
            if bandera==False:
                auxiliar=i
                bandera=True
            else:
                auxiliar=auxiliar+i
        else:
            lista.append(auxiliar)
            bandera=False
    return lista
```

```python
mi_lp_problema=pulp.LpProblem("Mi LP Problema", pulp.LpMinimize)

x11 = pulp.LpVariable('x11', lowBound=0, cat='Continuous')
x12 = pulp.LpVariable('x12', lowBound=0, cat='Continuous')
x13 = pulp.LpVariable('x13', lowBound=0, cat='Continuous')
x21 = pulp.LpVariable('x21', lowBound=0, cat='Continuous')
x22 = pulp.LpVariable('x22', lowBound=0, cat='Continuous')
x23 = pulp.LpVariable('x23', lowBound=0, cat='Continuous')


#Objective function
mi_lp_problema+=8*x11 +6*x12 +10*x13 +10*x21 +4*x22 +9*x23

#Consraints
mi_lp_problema += x11 + x12 + x13 ==2000
mi_lp_problema += x21 + x22 + x23 ==2500
mi_lp_problema += x11 + x21 ==1500
mi_lp_problema += x12 + x22 ==2000
mi_lp_problema += x13 + x23 ==1000
mi_lp_problema += x11>=0
mi_lp_problema += x12>=0
mi_lp_problema += x13>=0
mi_lp_problema += x21>=0
mi_lp_problema += x22>=0
mi_lp_problema += x23>=0
```
